```python
import re
import pandas as pd 
from pprint import pprint

from nltk.corpus import stopwords
stopwords = stopwords.words('spanish')

import pyLDAvis.gensim
from gensim.models import LdaModel
from gensim.corpora import Dictionary

import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
```

```python
# Cargar datos
path = 'C:/Users/ITHELATH/Documents/Laptop Col/dektop/NLP/Taller 8/reviews_vidjew_es.csv'
data = pd.read_csv(path)
data.head()
```

###  `[12 pts]` Punto 1: Hacer pre-procesamiento del texto

```python
def pre_procesado(texto):
    texto = texto.lower()
    texto = re.sub(r"[\W\d_]+", " ", texto)
    texto = [palabra for palabra in texto.split() if palabra not in stopwords]
    return texto
data['pp'] = data['review_body'].apply(lambda texto: pre_procesado(texto))
data.head()
```

###  `[13 pts]` Punto 2: Modelo de LDA

```python
# Crear una representación de los documentos en forma de diccionario
dic = Dictionary(data['pp'].values)

# Filtrar palabras muy frecuentes o infrecuentes
dic.filter_extremes(no_below= 5, no_above=0.7)

corpus = [dic.doc2bow(text) for text in data['pp'].values]

#Train the topic model
model = LdaModel(corpus=corpus, id2word=dic, num_topics=7, passes=100)
model.print_topics(num_words=15)
```

###  `[25 pts]` Punto 3: Visualización de LDA

```python
lda_display = pyLDAvis.gensim.prepare(model, corpus, dic, sort_topics=False)
pyLDAvis.display(lda_display)
```
