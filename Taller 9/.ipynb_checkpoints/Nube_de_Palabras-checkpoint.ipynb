```python
import numpy as np
from wordcloud import WordCloud
from wordcloud import ImageColorGenerator

from nltk.corpus import stopwords
stopwords = stopwords.words('spanish')

import matplotlib.pyplot as plt
import PIL.Image
from IPython.display import display
```

```python
with open("C:/Users/ITHELATH/Documents/Laptop Col/dektop/NLP/Taller 9/IHAVEADREAM.txt", encoding="utf-8") as archivo:
    data = archivo.read()
```

```python
wordcloud = WordCloud(width=800, height=400,
                     max_font_size= 150, max_words=250,
                     background_color='white',
                     colormap='gist_heat',
                     stopwords=stopwords).generate(data)

plt.figure(figsize=(10,8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```

```python
blackpower = PIL.Image.open("h.jpg")
display(blackpower)
```

```python
blackpow = np.array(blackpower)
```

```python
wordcloud = WordCloud(width=800, height=400,
                     max_font_size= 150, max_words=250,
                     background_color='black',
                     colormap='seismic',
                      mask=blackpow, contour_width=0.5, contour_color='white',
                     stopwords=stopwords).generate(data)

plt.figure(figsize=(10,8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
